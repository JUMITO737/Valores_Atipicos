```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel('Gastos y costos 20-23.xlsx',sheet_name = "2023", header=3)
df
```

```python
df.info()
```

```python
df.shape
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

### Valores nulos

```python
df1 =df.fillna(method="bfill")
df1['Poliza'] = df1['Poliza'].fillna(method="ffill")
df1['MP'] = df1['MP'].fillna(method="ffill")
df1['FP'] = df1['FP'].fillna(method="ffill")
df1 = df1.drop('Otros', axis=1)
df1
```

```python
valores_nulos=df1.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
df1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

### Desviación Estándar

```python
y=df1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
df2= df1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df2
```

```python
df3= df2.drop(['Fecha', 'Folio',	'UUID', 'RFC', 'Proveedor', 'TIPO GASTO', 'Descripción', 'MP', 'Tipo', 	'Status','Poliza'],axis = 1)
df3
```

```python
df3.info()
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df3.fillna(df3.mean(), inplace=True)
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
fig = plt.figure(figsize =(15, 8))
df3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

### Rango Intercuartílico

```python
y = df2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
df4 = df2.copy()
```

```python
df5 = df4.drop(['Fecha', 'Folio',	'UUID', 'RFC', 'Proveedor', 'TIPO GASTO', 'Descripción', 'MP', 'Tipo', 	'Status','Poliza'], axis=1)
df5
```

```python
valores_nulos=df5.isnull().sum()
valores_nulos
```

```python
df6 = df5.copy()
df6 = df6.fillna(round(df5.mean(),1))
df6
```

```python
fig = plt.figure(figsize =(15, 8))
df6.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show()
```

```python
indice=df['Fecha']
Datos_limpios = pd.concat([indice, df6], axis=1)
Datos_limpios
```
